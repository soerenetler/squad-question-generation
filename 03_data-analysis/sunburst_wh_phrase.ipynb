```python
import pandas as pd
from IPython.display import Image

from sunburst_utils import *
```

```python
df = pd.read_csv("random_train_question_features.csv")
```

```python
levels = ['WHP_token', 'WHP_type'] # levels used for the hierarchical chart
color_columns = 'count'
value_column = 'count'

df_all_trees = build_hierarchical_dataframe(df, levels, value_column, color_columns)
```

```python
import plotly.express as px
fig =px.sunburst(
    df_all_trees,
    names='name',
    ids='id',
    parents='parent',
    values='value',
    branchvalues='total',
    maxdepth=3,
    height=700
)

filename="images/questions/sunburst_wh_phrase.png"
fig.write_image(filename)
Image(filename)
```
